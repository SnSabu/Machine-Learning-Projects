# Webscraping Gas Prices

In this notebook, average gas prices from each of the input zip codes are scraped from the Autoblog website. Here the BeautifulSoup library is used to extract information from the html page.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#zipcodes = US_postal_codes["postal_code"].head(50).values.tolist()
zipcodes = [
'94591',
'94609',
'11040',
'17201',
'77375',
'85132',
'90010',
'75225',
'90814']

zip_price = []

all_gas_prices = [] # list to store all gas prices from all zipcodes

for zipcode in zipcodes:
    url = "https://www.autoblog.com/" + zipcode + "-gas-prices/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find(id="shops-map")
    gasprice = results.find_all("li", class_="slab price")
    values = [float(item.find("data")["value"]) for item in gasprice]
    Avg_Price = sum(values)/len(values)
    all_gas_prices.append(Avg_Price)
    zip_price.append((zipcode,Avg_Price))
zip_price_df = pd.DataFrame(zip_price,columns=['Zip Code', 'Gas Price'])


#print(all_gas_prices)

zip_price_df
```
